```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 11)
y = np.linspace(-1, 1, 11)

x, y = np.meshgrid(x, y)

z = np.exp(-(x ** 2 + y ** 2))

fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x, y, z, cmap="viridis",
                       linewidth=0, antialiased=False)

plt.show()
```

```python
import pandas as pd

data = pd.read_excel("datos3d.xlsx", sheet_name="Hoja 1")

data.head()
```

```python
x = data["x"]
y = data["y"]
z = data["z"]

X = []
Y = []
Z = []

for i in range(len(x)):
    xi = [xj for xj in x]
    yi = [y[i]] * len(x)
    zi = [zj for zj in z]
    X.append(xi)
    Y.append(yi)
    Z.append(zi)
    
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap="viridis",
                       linewidth=0, antialiased=False)

plt.show()
```
