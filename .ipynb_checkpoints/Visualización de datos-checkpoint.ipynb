```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5, 6]
y1 = [4, 3, 5, 2, 3, 1]

x2 = [1, 2, 3, 4, 5, 6]
y2 = [3, 7, 3, 2, 1, 2]

x3 = [1, 2, 3, 4, 5, 6]
y3 = [3, 4, 6, 2, 2, 3]

plt.plot(x1, y1, "s-", color="red")
plt.plot(x2, y2, "o--", color="blue", lw=3) # lw <-> linewidth
plt.plot(x3, y3, "s", color="gray", ms=10, mfc="magenta")

# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

plt.show()
```
